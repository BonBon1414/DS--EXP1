```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/health.csv')
df
```

Encoding Categorical Variables

```python
print(df.columns)
```

One Hot Encoding

```python
df = pd.get_dummies(df, columns=['Gender'])
df
```

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Year']=le.fit_transform(df['Year'])
df
```

Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
size_mapping = {'tertiary':0, 'secondary':1, 'unknown':2}
df['Race/ Ethnicity']=df['Race/ Ethnicity'].map(size_mapping)
df
```

Frequency Encoding

```python
frequency_encoding = df['Place'].value_counts().to_dict()
df['Place'] =df['Place'].map(frequency_encoding)
df
```

Target Encoding

```python
mean_encoded = df.groupby('Value')['Place'].mean().to_dict()
df['Value'] = df['Value'].map(mean_encoded)
df
```

Hashing Encoding

```python
df['Indicator Category'] = df['Indicator Category'].astype(str)
df
```

```python
import pandas as pd
from sklearn.feature_extraction import FeatureHasher

data = {'Indicator Category': ['HIV/AIDS', 'Cancer', 'Maternal and Child Health', 'HIV/AIDS', 'Life Expectancy and Death Rate (Overall)', None],
        'Year': [10, 8, 9, 10, 9, 8]}
df = pd.DataFrame(data)
df
```

```python
df['Indicator Category'] = df['Indicator Category'].fillna('Unknown')
data_dicts = df.apply(lambda row: {'Indicator Category': row['Indicator Category'], 'Year': row['Year']}, axis=1).tolist()
```

```python
hasher = FeatureHasher(input_type='dict', n_features=8)
hashed_features = hasher.transform(data_dicts)
df_hashed = pd.DataFrame(hashed_features.toarray(), columns=[f'feature_{i}' for i in range(8)])
print(df_hashed)
```
